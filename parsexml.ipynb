```python
import pandas as pd
import numpy as np
import os
from dict2xml import dict2xml 

df = pd.read_csv('gland_proposal.csv')

for i in range(len(df)):
    test =df.iloc[i].tolist()
    d = {'folder': 'Annotations',
     'filename': test[0],
     'path': os.path.join('images',test[0]),
     'source': {'database': 'Unknown'},
     'size': {'width':test[1],
              'height': test[2],
              'depth': 1
             },
     'segmented':0,
     'object': {'name': 'prostate',
                'pose':'Unspecified',
                'truncated': 0,
                'difficult': 0,
                'bndbox':{'xmin':test[4],
                          'ymin':test[5],
                          'xmax':test[6],
                          'ymax':test[7],
                         }
               }
    
    }
    xml = dict2xml(d) 
    xmlfile = open(os.path.join('Annotations',test[0][:-4]+".xml"), "w") 
    xmlfile.write(xml) 
    xmlfile.close()
```
